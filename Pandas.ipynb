# **In Python , when you see the word of drop , it means python deletes some part of the data**
### *The function Drop is for deleting a part of the Dataframe*

# **in the argument of drop, the meaning of axis=0 means delete row with the mentioned index name and axis = 1 means delete column with the mentioned column name**




# If we use the function drop , this function does not delete columns and rows permanentely. So we have to define a new dataframe and we assign it to dropped dataFranme like the below example


```python
my_df
df_dropped = my_df.drop(index='row2',columns='Var5') # index and columns are arguments for function drop
df_dropped
```

```python
my_df
```

##1- The argument of inplace if is True means that you keep the change permanentely and if it is False , it is temproral change
##2- if you do not mention an argument in the function, Python considers its default value which in the case of in place is False.

```python
my_df.drop('row2',axis=0,inplace=True) # the argument of inplace if is True means that you keep the change permanentely and if it is False , it is temproral change
# if you do not mention an argument in the function, Python considers its default value which in the case of in place is False.
```

```python
my_df
```

### **Remember that the drop() function returns a new DataFrame by default, leaving the original DataFrame unchanged. If you want to modify the original DataFrame in-place, you can use the inplace=True argument:**

```python
my_df.drop('row3',axis=0,inplace=True)
my_df
```

----
### **1- Reads the file off disk (or steam)**

*Scan every cell in the file and, by default, treat empty fields (and common placeholders like NA, n/a, blank strings, etc.) as NaN.*
