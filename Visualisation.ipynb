## **Bar Chart**


---


The best tool for comparison and visualization of cathegorical paramaters

```python

# Look at index 4 and 6, which demonstrate overlapping cases.
# red bar
t1 = [1, 3, 4, 5, 6, 7, 9]
tt1 = [4, 7, 8, 4, 7, 8, 3]

# green bar
t2 = [2, 4, 6, 8, 10]
tt2 = [5, 6, 2, 6, 2]

# Colors: https://matplotlib.org/api/colors_api.html

plt.bar(t1, tt1, label="Red Bar", color='red')

plt.bar(t2, tt2, label="Green Bar", color='green')


plt.xlabel('group')
plt.ylabel("bar height")
plt.title("Bar Chart Example")
plt.legend() # this code is for putting legend on top of the diagram.
plt.show()
```

### ***Histogram Plot***

# Histogram Helps to Get Idea About Distribution
always has one variable and another axis is  number of observation of each amount of variable.

Histogram shows the type distribution of data , whether it is normal or abnormal , moreover it shows the amount of standard deviation sd.

We usually use histogram for discrete numbers (like counts: 1,2,3,4,...) , but if we have continuous numbers then , it group the numbers into ranges out of the numbers.

###1- The blue distribution is the most centered and narrow.
Very consistent values and tightly clustered around the mean.

###2- red distribution is shifted to the right, data is larger in value.
Moderate variation and higher average value


###3- The green distribution is slightly to the left and more spread out.
Higher Variation and lower value

This list is not directly used in the plot but define the context of population of IRAN.

```python
city_name = ['Tehran', 'Mashhad', 'Isfahan', 'Karaj', 'Tabriz', 'Shiraz']
city_pop = [7153309, 2307177, 1947164, 1448075, 1424641, 1249942]
```

```python
import matplotlib.pyplot as plt #library for all types of plots and graphs including histograms.
plt.figure(figsize=(6,6), dpi=100)
plt.bar(city_name, city_pop)
plt.show()

plt.hist(city_pop,color='skyblue')
plt.xlabel('population')
plt.ylabel('City Number: Frequency') # describe how many cities fall into each bin=group
plt.grid(True, linestyle='--', alpha=0.5) # drawing the grid
plt.show()
```

```python
import matplotlib.pyplot as plt

city_name = ['Tehran', 'Mashhad', 'Isfahan', 'Karaj', 'Tabriz', 'Shiraz']
city_pop = [7153309, 2307177, 1947164, 1448075, 1424641, 1249942]

plt.figure(figsize=(6,4), dpi=100)

# Draw the histogram
plt.hist(city_pop, bins=5, color='skyblue', edgecolor='black', alpha=0.7)

# Labels
plt.xlabel('Population')
plt.ylabel('Number of Cities (Frequency)')
plt.title('City Population Distribution')

plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

# **Box Plot**


---


**In Box Plot , we can show the median, 25 percentile (Q1) , 75 percentile (Q3) , upper extreme and lower extreme . By this plot , we can idetify the outliers of the data distribution including upper and lower outliers.** **IQR = Q3-Q1**


### ***The size of the box is important becasue it shows how data is distributed and which range has most distribuation of data***
*italicized text*

```python
import seaborn as sb
import matplotlib.pyplot as plt

sb.boxplot(y='bp_after', data=example3) #showing the box p;ot of the blood pressure after the treatment
plt.show()
```

# ***Strip Plot***


---


## plots the distribution of variables for each category as individual datapoints

```python
import pandas as pd #importing pandas library for data analytics
import seaborn as sb #importing seaborn for creation statistical graphics
```

```python
smartphones=pd.read_csv('/content/smartphones.csv')
smartphones
```

```python
smartphones.info()
```

```python
smartphones.describe()
```

```python
plt.figure(figsize=(13,8))
sb.stripplot(x='OS', y='Capacity', data=smartphones, hue='Ram', size=30)
plt.show()
```

### ***Swarm Plot***

## This function is similar to stripplot(), but the points are adjusted (only along the categorical axis)
## **so that they don’t overlap on each other**

```python
plt.figure(figsize=(13,8))
sb.swarmplot(x='OS', y='Capacity', data=smartphones, hue='Ram', size=30)
plt.show()
```

```python
plt.figure(figsize=(13,8))
sb.swarmplot(x='OS', y='Capacity', data=smartphones, hue='Company', size=30) # a swarm plot showing different companies with their capacity and their software which don't overlap each other
plt.show()
```

# ***Pair Plot***


---


###This plot is good practical plot to undestand the corellations between variable pairs

Plot pairwise relationships in a dataset.

By default, this function will create a grid of Axes such that each numeric variable in ``data`` will by shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.


It is also possible to show a subset of variables or plot different variables on the rows and columns.

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt


salaries = pd.read_csv('/content/Salaries.csv')
salaries
```

```python
sb.set_style('darkgrid') # this the command for changing backgroun of the graph

sb.pairplot(salaries, hue='rank',plot_kws={'s':100}) # creating the pair plot based on the dataframe of salaries, data is explained basd on the rank type "professors"
plt.show()
```
