# **Clustering**

###- A type of unsupervised learning

###- The **Goal** is to group data points into “clusters” so that items in the same cluster are as similar as possible, and items in different clusters are as dissimilar as possible.

###- **No labels are provided**; the algorithm finds structure purely from the input features.



---

# **K-Means**

###- One of the most popular clustering algorithms
###- It is one of the simplest and fastest unsupervised ML methods, specifically for large datasets.
###- Each cluster is described by its centroid (average data point) and causes its **easy interpretation**.



---

# **Disadvantages**

###- We must know how many clusters we want.

###- It works best when clusters are in similar size, and when our dataset has weird shapes, K-Means may spplit and merge them incorrectly.

###- It is sensitive to initialized centroids, and bad initialization causes inappropriate final solutions.
###- Outliers affect Centroids and they can pull toward itself.



---

# **4 Main Steps for K-Mean Algorithm:**



##**1- Initialization**

####First, "***N*** " nodes will be considered as datapoints, and then randomly place "***k*** " clusters as crosses among those nodes will be created.

image.png


##**2- Assignment Step**

###For each node, the nearest cross (cluster center) will be found, then this center will be assigned to the nearest cross.  

### **Note:** The distance calculation will be calculated by ***Euclidean Method***.

image.png

## **3- Update Step**

###For each cross (cluster), the average position of all nodes (centroid) will be assigned to the cluster and then, the cross will move to the updated centroid.

image.png

##**4- Repeat**

###Return to Step 2 (Assignment) and continue until clusters no longer change (or move very little).

image.png

# **Evaluation of the Precision**

##In ***supervised*** learning, we typically measure performance by comparing predicted labels to true labels (e.g., accuracy, precision/recall on a held-out test set).
## In ***pure unsupervised*** learning there are no “true” labels, so you can’t compute precision or recall in the same way. Instead, for K-Means clustering one of the most common internal metrics is "***inertia*** ".


# **inertia = ∑ₖ ∑_{xᵢ ∈ cluster k} ||xᵢ – μₖ||²**


##Inertia is literally the sum of squared Euclidean distances from each point to its cluster centroid.

##Lower inertia means points are, on average, closer to their centroids (more “compact” clusters).



---


## Our purpose is to find the best number for K which shows the minimum numbers of clusters, which are the best representative of their members.

## ***(Lower K with Lower Inertia = Best Case)***

```python
kmn.inertia_ # it gives you the intertia ( showing the differences among objects inside one cluster)
```



---

# Determining the "Right" K

## - In K-Means, you must pick the number of clusters (K) before fitting. If K is too small, very different data points get forced together; if K is too large, you may be splitting natural groups into tiny fragments.



---

# Definition of the “Elbow”

## As K increases, inertia always decreases (because adding more centroids can only make clusters tighter).

## At first, going from K=1→2→3 often yields a large drop in inertia—each new cluster explains a lot of variance.

## Beyond some point (the “elbow”), adding more clusters yields only small, diminishing reductions in inertia.

## The “elbow” is the value of K at which the plot’s curvature changes from steep to gradual—a natural trade-off between model complexity (more clusters) and compactness.

image.png

```python
inertia_list = []

# make a list which we add the caluculation of inertial in each loop to this list
for k in np.arange(1,40): # the structure of loop should be consistent and the next lines should be after for .
    kmn =KMeans(n_clusters = k)
    kmn.fit(df)
    inertia_list.append(kmn.inertia_)
inertia_list
```

```python
plt.plot(np.arange(1,40),inertia_list,'o-',c='green')
plt.xlabel('number of clusters')
plt.ylabel('Inertia')
plt.show()
```
