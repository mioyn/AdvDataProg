## **For machine Learning, there are 3 steps:**


###**1.   Creating the Model**
*choosing your algorithm and telling your computer which one to use. It is the initial setupchoosing and configuring the toolbefore you train it on your data.*

###**2.   Fitting the Data**
*Give the model your training data so it learns how it can make the prediction.*

*Your machine makes a guess about new data it has never seen before.*







---


##**1- Supervised Machine Learning**:


*   involves training a model on a labeled dataset


*   each training example consists of input features (known as **X**) and a known target or label (known as **Y**).


*   The model learns a mapping from inputs to outputs so it can predict labels on new, unseen data.
















---


## **Key components of Supervised Machine Learning:**


###**1. Features (X):**

  *Input variables (e.g., pixel values of an image, measurements of a house).*

###**2. Labels (y):**

  *Known outputs (e.g., cat vs. dog, house price).*

###**3. Model:**

  *An algorithm (e.g., decision tree, linear regression, neural network) that learns parameters to map X  y.*

###**4. Training:**

  *Adjusting model parameters to minimize a loss function on the labeled data.*

###**5. Evaluation:**

  *Measuring performance on a held-out test set using metrics like accuracy (classification), mean squared error (regression), precision, recall, etc.*

##***KNN: K-Nearest Neighbors***

###one of the simplest, yet surprisingly effective, non-parametric, example-based learning algorithms in supervised machine learning.



#***Core Idea***
* **Instance-based (lazy) learning**


KNN defers all computation until prediction time. It simply remembers the entire training set.

 * **Local decision rule**

To predict the label of a new point, KNN looks at the K closest stored examples and lets them vote.





---

###1. Choose **K** as the number of neighbors.

###2. Pick a distance metric, most commonly Euclidean distance. (Manhattan, Euclidean, Minkowski)

```python
from sklearn.neighbors import KNeighborsClassifier # First step: creat model
knn = KNeighborsClassifier(n_neighbors=18, metric='minkowski',p=2) # creat model because P= 2 then the method for distance calculation is Euclidiean Method
x = iris.data
y = iris.target
knn.fit(x, y) # Second step:Fitting the model
```

# **2- Train and Test**

image.png

## **Test_size:**
* Specifies how much the overall dataset is allocated for testing.

* Accepts a float between 0 and 1, representing the fraction of samples to use as the test set.

* In the **below** example, `test_size=0.3` reserves 30% of the data for testing and leaves the remaining 70% for training.



##**Random State:**

* Provide a set of fixed seed for the random number generator

* Ensuring the train/test spilit stays equal across multiple runs.

* Ensuring the consisitent results for analysis and debugging.

* Specific value of `random_state` locks the exat partition of data each time the code is running.

## **Stratify**

* Keeps the same mix of classes in test and train process.

* prevents ending up with 0 examples of a rare class in the test set.

* Enabling reliable and fair evaluation of the Model's performance on every class.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3,stratify=iris.target,random_state=42)
```

#**3- Prediction**

###In Prediction Process, Features of the new object are given to the fitted KNN and the KNN Model identifies the class of the object.

## **Note**: we use ***Train Sample*** for fitting the data and ***Test sample*** for Prediction.

```python
knn = KNeighborsClassifier(n_neighbors=18)
knn.fit(x_train, y_train) # fitting the model based on the train samples
y_predict = knn.predict(x_test) # predict model based on test sample
y_predict # showing the labels of test data
```

## **Over Fitting and Under Fitting**

####**Over fitting**

means the accuracy of model for train sample is very nice and perfect near to 1.0 but for test sample the perfomace ( accuracy) is very low which shows that we can not generalize the model for different samples.

####**Under Fitting**

This means that the model is too simple to classify the objects correctly, and we can not generalize it to all samples.

###***The best situation is the same accuracy and high enough ( for example around 70 to 95 % accuracy) between train and test samples.***

```python
neighbors = np.arange(1, 30)

train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
  knn_model = KNeighborsClassifier(n_neighbors=k)
  knn_model.fit(x_train, y_train)
  train_accuracy[i] = knn_model.score(x_train, y_train)
  test_accuracy[i] = knn_model.score(x_test, y_test)

plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```
